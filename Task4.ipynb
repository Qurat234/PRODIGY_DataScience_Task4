## Analyze and visualize sentiment patterns in social media data to understand public opinion and attitudes towards specific topics or brands.
 DataSet Link : https://www.kaggle.com/datasets/jp797498e/twitter-entity-sentiment-analysis

```python
# Tools & Libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('twitter_training.csv')
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
column_name=['tweetID','entity','sentiment','tweet_content']
df.columns=column_name
df
```

* show the names of the columns in our dataframe:

```python
df.dtypes # info of dtypes
```

```python
df.isna().sum() 
```

```python
df.dropna(inplace=True) # Remove the all null values
df.isna().sum()
```

```python
#unneeded columns:
df.drop(columns=['tweetID','tweet_content'],inplace=True)
df.head()
```

```python
df.shape
```

```python
entity_content=df['entity'].value_counts()

entity_content.plot(kind='pie', autopct='%1.1f%%', figsize=(10, 12))
plt.title('Distribution of entities')
plt.show()
```

```python
# Count the occurrences of each sentiment label
sentiment_counts = df['sentiment'].value_counts()
colors = ['skyblue', 'salmon', 'blue', 'red']
# Plot the sentiment distribution
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values,hue=sentiment_counts.index,palette=colors)
plt.title('Sentiment Distribution for Borderlands')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

* at the above figure we can see the number of the negative reactions are above to 20000
the min number of the reactions are the number of the irrelevant reactions are less tan 15000

```python
plt.figure(figsize=(16, 6)) 
sns.countplot(x='entity', data=df, hue='sentiment', edgecolor='black', saturation=0.75) 
plt.title('Distribution of Entity by Sentiments')  
plt.legend(title='Sentiment') 
plt.xticks(rotation=90);
```
